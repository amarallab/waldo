```python
ex_id = '20130318_131111'
dataset = 'N2_aging'
data_type = 'length_mm'
#data_type = 'curve_w'
for blob, times, data in pe.iter_through_worms(ex_id, data_type):
    ppl.plot(times, data)
plt.show()
    
```

```python
ids, times, datum = zip(*[i for i in pe.iter_through_worms(ex_id, data_type)])
starts, ends, dur, mids = [], [], [], []
for t in list(times):
    starts.append(t[0])
    ends.append(t[-1])
    dur.append(t[-1] - t[0])
    mids.append((t[-1] + t[0])/2.0)

zipdata = zip(datum, starts, ends, dur)
```

```python
fig, ax = plt.subplots(2,1)
sort_key= 1 # sort key
data, starts, ends, durs = zip(*sorted(zipdata, key=lambda x: x[sort_key]))
ppl.plot(ax[0], starts, label='start')
ppl.plot(ax[0], ends, label='end')
ppl.plot(ax[0], durs, label='dur')
ppl.boxplot(ax[1], data)
ax[0].set_title('sorted by start time')
ax[1].set_xlabel('blob tracks')
ax[0].set_ylabel('time (s)')
ax[1].set_ylabel(data_type)
ax[0].legend(loc='upper left')

plt.show()
```

```python

first = 0
last = 3600

N_time = int(last - first)
N_blobs = len(times)

sort_key= 3 # sort key
data, starts, ends, durs = zip(*sorted(zipdata, key=lambda x: x[sort_key]))
is_there = np.zeros((N_blobs, N_time))
for i, (start, end) in enumerate(zip(starts, ends)):
    is_there[i][int(start):int(end)] = 1

fig, ax = plt.subplots()
ppl.pcolormesh(fig, ax, is_there)
ax.set_title('when blobs exist (sorted by duration)')
ax.set_ylabel('track')
ax.set_xlabel('time (s)')
plt.show()
```
