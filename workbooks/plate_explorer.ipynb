```python
# Function definitions
def sort_zipdata(zipdata, N_data_col=2, sort_index='dur', sort_labels=[], verbose=False):
    '''
    returns the zipdata info sorted by a chosen sort_index
    
    params
    --------
    zipdata: (list of lists)
        each list in zipdata follows this pattern
           [id, data_col1, data_col2, ... data_colN, sort_col1, ...sort_colN]
        where:
        id - an identifying string. (ex. blob_ids)
        data_columns - anything you want to sort. (ex. each value is a timeseries for a blob_id)
        sort_columns - single numerical values. (ex. start time or standard deviation)
    N_data_col: (int)
        the number of data columns being sorted
    sort_index: (str)
        the name of the value you want to sort on
    sort_labels: (list of str)
        the list containing the names of every sort_col
    verbose: (bool)
        if true, prints info
        
    returns:
    ------
    id_col: (list)
        sorted list of ids
    data_col: (tuple)
        contains sorted versions of all data columns
    sort_cols: (tuple)
        contains sorted versions of all sort_columns
    
    '''
    assert sort_index in sort_labels, 'sort_index must be present in the sort_labels list'
    index_col = sort_labels.index(sort_index) + N_data_col + 1
    unzipped = zip(*sorted(zipdata, key=lambda x: x[index_col]))
    id_col, data_cols, sort_cols = unzipped[0], unzipped[1:N_data_col+1], unzipped[N_data_col+1:]
    if verbose:
        print len(data_cols), 'data columns'
        print len(sort_cols), 'sort columns:', sort_labels
    assert len(sort_cols) == len(sort_labels), 'sort labels must have same len as sorting part of zipdata'
    return id_col, data_cols, sort_cols
```

```python
def plot_sorted_boxplot(ax1, ax2, zipdata, sort_index='start', sort_labels=[]):
    """
    
    """
    id_col, data_cols, sort_cols = sort_zipdata(zipdata, sort_index=sort_index, sort_labels=sort_labels)
    times, data = data_cols
    for d, l in zip(sort_cols, sort_labels):
        ppl.plot(ax1, d, label=l, linestyle='steps')
        
    ax1.legend(loc='upper left')
    ax1.set_ylabel('time (s)')
    ax1.yaxis.set_ticks_position('both')
    ppl.boxplot(ax2, data)
    ax2.set_xlabel('blob tracks')
    ax2.yaxis.set_ticks_position('both')

fig, ax = plt.subplots(2,1) 
plot_sorted_boxplot(ax[0], ax[1], zipdata, sort_index='start', sort_labels=['start', 'end', 'dur'])
ax[0].set_title('{eid} {dt} (sorted by start time)'.format(eid=ex_id, dt=data_type))
ax[1].set_label(data_type)

fig, ax = plt.subplots(2,1) 
plot_sorted_boxplot(ax[0], ax[1], zipdata, sort_index='dur', sort_labels=['start', 'end', 'dur'])
ax[0].set_title('{eid} {dt} (sorted by duration)'.format(eid=ex_id, dt=data_type))
ax[1].set_label(data_type)
plt.show()

fig, ax = plt.subplots(2,1) 
plot_sorted_boxplot(ax[0], ax[1], zipdata, sort_index='end', sort_labels=['start', 'end', 'dur'])
ax[0].set_title('{eid} {dt} (sorted by end time)'.format(eid=ex_id, dt=data_type))
ax[1].set_label(data_type)
plt.show()
```

```python
def exist_plot(fig, ax, times, first=0, last=3600):

    N_time = int(last - first)
    N_blobs = len(times)
    starts = [min(t) for t in times]
    ends = [max(t) for t in times]
    is_there = np.zeros((N_blobs, N_time))
    for i, (start, end) in enumerate(zip(starts, ends)):
        is_there[i][int(start):int(end)] = 1
        
    ppl.pcolormesh(fig, ax, is_there)
    ax.set_ylabel('track')
    ax.set_xlabel('time (s)')
        
sort_index = 'start'
id_col, data_cols, sort_cols = sort_zipdata(zipdata, sort_index=sort_index, sort_labels=sort_labels)  
times, data = data_cols
fig, ax = plt.subplots()
exist_plot(fig, ax, times)
ax.set_title('{eid} (sorted by {i})'.format(eid=ex_id, i=sort_index))
plt.show()

sort_index = 'dur'
id_col, data_cols, sort_cols = sort_zipdata(zipdata2, sort_index=sort_index, sort_labels=sort_labels2)
times, data = data_cols
fig, ax = plt.subplots()
exist_plot(fig, ax, times)
ax.set_title('{eid} (sorted by {i})'.format(eid=ex_id, i=sort_index))
plt.show()
```
