```python
def exponential_fit(s, bins=60):
    s = np.array(s)
    loc, scale = stats.expon.fit(s)
    return loc, scale

def plot_exponential_fit(s, bins=60):
    s = np.array(s)
    loc, scale = stats.expon.fit(s)
    dist = stats.expon(loc=loc, scale=scale)
    #print loc, scale
    x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), bins)
    #print len(x)
    fig, ax = plt.subplots(1, 2)
    ax[0].hist(s, bins=x, normed=True)
    ax[0].plot(x, dist.pdf(x),'r-', lw=5, alpha=0.6, label='expon pdf', color='black')
    ax[1].hist(s, bins=x, normed=True)
    ax[1].plot(x, dist.pdf(x),'r-', lw=5, alpha=0.6, label='expon pdf', color='black')
    ax[1].set_yscale('log')
    ax[1].set_title('scale = {:0.1e}'.format(scale))

    return ax[0]
    
ax = plot_exponential_fit(sdf['speed'])
ax.set_title('full')
plt.savefig('{bid}-full-hist.pdf'.format(bid=bid))

plt.show()

ax = plot_exponential_fit(df1['speed'])
ax.set_title('before 20 min')
plt.savefig('{bid}-before-hist.pdf'.format(bid=bid))
plt.show()

ax = plot_exponential_fit(df2['speed'])
ax.set_title('after 20 min')
plt.savefig('{bid}-after-hist.pdf'.format(bid=bid))
plt.show()

ax = plot_exponential_fit(df2[(df2['time'] > 20*60) & (df2['time'] < 35*60)]['speed'])
ax.set_title('20-35')
plt.savefig('{bid}-after-hist.pdf'.format(bid=bid))
plt.show()

ax = plot_exponential_fit(df2[(df2['time'] > 35*60) & (df2['time'] < 50*60)]['speed'])
ax.set_title('35-50')
plt.savefig('{bid}-after-hist.pdf'.format(bid=bid))
plt.show()

f, ax = plt.subplots()
starts = np.linspace(20, 40)
win_len = 10
fits = []
for start in starts:
    data = df2[(df2['time'] > start*60) & (df2['time'] < (start+win_len)*60)]['speed']
    fits.append(exponential_fit(data)[1])
    
plt.plot(starts, fits)
plt.title('scale factor in sliding window')
plt.xlabel('window start (min)')
plt.savefig('{bid}-sliding-window.pdf'.format(bid=bid))


f, ax = plt.subplots()
starts = np.linspace(20, 40)
win_len = 10
fits = []
start = 20
t0 = 5
tN = int(max(df2['minutes'])) - 20
wins = np.linspace(5, tN, tN-t0)

for win_len in wins:
    data = df2[(df2['time'] > start*60) & (df2['time'] < (start+win_len)*60)]['speed']
    fits.append(exponential_fit(data)[1])
    
plt.plot(wins, fits)
plt.title('scale factor in sliding window')
plt.xlabel('window size (min)')
plt.savefig('{bid}-window-size.pdf'.format(bid=bid))

```

```python
from pandas.tools.plotting import autocorrelation_plot
    
def autocorr(x, t=1):
    return np.corrcoef(numpy.array([x[0:len(x)-t], x[t:len(x)]]))[0, 1]
```

```python
def acf_last_significant(df, col='speed', plot=False, ax=None):
    x = df[col]
    lags = np.arange(len(x))
    nlags = len(lags) - 1
    
    fft = True
    unbiased = False
    alpha = 0.05
    acf_x, confint = acf(x, nlags=nlags, alpha=alpha, fft=fft,
                         unbiased=unbiased)

    confint = confint - confint.mean(1)[:,None]
    top_conf = confint[:,1]
    bottom_conf = confint[:,0]
    for i, (ac, t) in enumerate(zip(acf_x, top_conf)):
        if ac <= np.abs(t):
            break
            
    if plot:
        if ax is None:
            fig, ax = plt.subplots()    
        #ax.margins(.01)
        ax.plot(lags, acf_x, color='black')
        ax.fill_between(lags, confint[:,0], confint[:,1], alpha=.25)
        ax.plot([lags[i-1]], [ac], 'o', color='red')
        ax.set_title("Autocorrelation")
        ax.set_ylim([min(acf_x), 1.1])

    return lags[i - 1]
```
