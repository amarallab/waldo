```python
def plot_property(bin_df, key='p-ks', xkey='dur', logy=False, ylim=None):

    fig, ax = plt.subplots()
    if logy:
        plt.semilogy()
    
    piv = pd.pivot_table(bin_df, values=key, rows=xkey)
    bin_size = bin_df[xkey]
    p = bin_df[key]
    ax.plot(bin_size, p, ls='None', marker='.')
    
    x_piv = np.array(piv.index)[:]
    average = np.array(piv)[:]
    ax.plot(x_piv, average)
    if xkey == 'dur':
        ax.set_xlabel('bin size (min)')
    elif xkey == 'rel-bin-size':
        ax.set_xlabel('fraction of track length')
    if ylim is not None:
        ax.set_ylim(ylim)
    ax.set_ylabel(key)
    ax.set_xlim([min(bin_size) - 1 , max(bin_size) + 1])
    #ax.set_ylim([-0.05, 0.1])

def plot_reject_fraction(bin_df, xkey='dur'):
    alpha = 0.05

    bins = []
    counts = []
    fraction_less_alpha = []
    for dur, dur_df in bin_df.groupby(xkey):
        bins.append(dur)
        count = len(dur_df)
        counts.append(count)
        reject_count = len(dur_df[dur_df['p-ks'] < alpha])
        fraction_less_alpha.append(float(reject_count) / float(count))


    fig, ax = plt.subplots()
    ax.plot(bins, counts, 'o-') #, label='counts')
    ax.set_ylabel('bin count')
    ax.set_xlabel('bin size (min)')
    ax.set_xlim([min(bins) - 1 , max(bins) + 1])
    ax.set_ylim([0 , max(counts) + 1])

    fig, ax = plt.subplots()
    ax.plot(bins, fraction_less_alpha, 'o-') #, label='reject fraction')
    print fraction_less_alpha
    ax.set_ylim([0, 1.1])
    ax.set_ylabel('fraction rejected')
    ax.set_xlabel('bin size (min)')
    ax.set_xlim([min(bins) - 1 , max(bins) + 1])

def describe_bin_df(bin_df):
    unique_blobs = list(set(bin_df['id']))
    unique_recordings = list(set(bin_df['eid']))

    print 'start position:', min(bin_df['start-pos']), 'minutes'
    print len(unique_blobs), 'unique tracks'
    print len(unique_recordings), 'unique recordings'
    print len(bin_df), 'total bins'

def pick_appropriate_subsection(full_bin_df, start_pos=0):
    assert start_pos in [0, 10, 20, 30], 'start pos must be 0, 10, 20 or 30'
    
    df = full_bin_df
    if start_pos == 0:
        # return full tracks
        return df[df['is-cropped'] == False]
    
    else:
        # condition 1
        bdf1 = df[(df['is-cropped'] == False)     # bin is not cropped
                  &                               # and
                  (df['start-pos'] >= start_pos)] # starts later than cuttoff
        
        # condition 2
        bdf2 = df[(df['is-cropped'] == True)      # bin is cropped
                  &                               # and
                  (df['start-pos'] == start_pos)] # starts at cuttoff   
        
        return pd.concat([bdf1, bdf2], axis=0)
```

```python
def rejections_by_track(df, bins=15):
    d = df[['id', 'p-ks']]
    df_reject = d[d['p-ks'] < 0.05].groupby('id').count()
    df_total = d.groupby('id').count()

    d_test = pd.DataFrame(index=df_reject.index)
    d_test['reject'] = df_reject['p-ks']
    d_test['total'] = df_total['p-ks']
    d_test['fraction-rejected'] = d_test['reject'] / d_test['total']

    fr = d_test['fraction-rejected']
    kernel = stats.gaussian_kde(fr, 0.25 )
    kx = np.linspace(0, 1, 1000)
    ky = kernel.evaluate(kx)

    fig, ax = plt.subplots()
    ax = d_test['fraction-rejected'].hist(bins=bins)
    ax.plot(kx, ky)
    ax.set_xlabel('fraction of total bins rejected')
    ax.set_ylabel('number of tracks')
    plt.show()

    
def bins_by_track(df):
    
    d = df[['id', 'p-ks']]
    df_total = d.groupby('id').count()
    #print df_total
    mn = min(df_total['id'])
    mx = max(df_total['id'])
    #print mn, mx
    bins = np.arange(mn-0.5, mx+1.5)
    #print bins
    #print df_total
    
    bc = df_total['id']
    #kernel = stats.gaussian_kde(bc, 0.25 )
    #kx = np.linspace(, 1, 1000)
    #ky = kernel.evaluate(kx)

    fig, ax = plt.subplots()
    ax = df_total['id'].hist(bins=bins)
    ax.set_xlim([mn - 0.5, mx + 0.5])
    ax.set_xlabel('total number of bins for track')
    ax.set_ylabel('number of tracks')
    plt.show()
    
#rejections_by_track(df)
#bins_by_track(df)
```
