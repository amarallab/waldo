```python
def autocorr_time_to_zero(df, nan_is_max=True, plot=False):
    s = df['speed']
    
    #     result = np.correlate(s, s, 'full')
    #     ac = result[result.size/2:]

    #     fig, ax = plt.subplots(2,1, sharex=True)
    #     ax[0].plot(s)
    #     ax[0].set_ylabel('speed')
    #     ax[1].plot(ac)
    #     ax[1].set_ylabel('AC')
    #     ax[0].set_xlim([0,len(s)])
    #     plt.show()

    def autocorr(x, t=1):
        return np.corrcoef(numpy.array([x[0:len(x)-t], x[t:len(x)]]))[0, 1]

    ac = np.array([autocorr(s, t=dt) for dt in range(len(s) - 5)])
    below_zero = np.where(ac <= 0.0)[0]
    try: 
        first_zero = below_zero[0]
    except IndexError:
        if nan_is_max:
            return len(s)
        else:
            return np.nan
        
    if plot:
        fig, ax = plt.subplots()
        ax.plot(ac)
        ax.plot([first_zero, first_zero], [-1, 1], 'black')
        plt.show()
    
    return first_zero
    #pd.tools.plotting.autocorrelation_plot(df['speed'])
    #     plt.show()
```

```python
def calculate_centroid_properties(df):
    
    s = np.array(df['speed'])
    loc, scale = stats.expon.fit(s)

    
    prop = {
            'mean': np.mean(s),
            'std': np.std(s),
            'skew': stats.skew(s),
            'kurt': stats.kurtosis(s),

            '50%': np.median(s),
            '25%': np.percentile(s, 25),
            '75%': np.percentile(s, 75),
            'inter_quartile': np.percentile(s, 75) - np.percentile(s, 25),
            'expon_scale': scale,
            }

    return prop


def pull_track_properties(eid, start_min=20, min_dur=20):
    """
    Applies calculate_centroid_properties() to every track
    in the dataframe that is longer than 'min_dur' minutes long
    after the first 'start_min' minutes of recording have been
    cut off.
    
    params
    -----
    start_min: (int)
        minute
    min_dur: (int)
        minute
    """
    sw = SpeedWriter(eid)
    typical_bl = sw._experiment.typical_bodylength

    wr = WormWriter(eid)
    bids = wr.blobs('speed_files')
    
    bins = [5,10]
    all_tracks = []
    for bid in bids[:]:
        speed_df = wr.load(bid, data_id='speed_files')
        speed_df['time'] = speed_df['time'] / 60.0 # convert to min
        speed_df = speed_df[speed_df['time'] > start_min]
        
        if len(speed_df) < 60:
            continue
            
        track_dur = max(speed_df['time']) - min(speed_df['time'])
        if track_dur < min_dur:
            continue
            
        #print bid
        unique_id = '{e}-{b}'.format(e=eid, b=bid)        

        start_time = min(speed_df['time'])
        # Calculate full track properties
        #return None
        description = calculate_centroid_properties(speed_df)
        bl_sprint = fastest_bl_sprint(speed_df, typical_bl) *60 
        
        description.update({'bid': bid,
                            'eid': eid,
                            'bin-size': track_dur,
                            'id': unique_id,
                            'bl-sprint': fastest_bl_sprint(speed_df, typical_bl),
                            'bl-5-sprint': fastest_bl_sprint(speed_df, typical_bl*5), 
                            'ac-time-to-0': autocorr_time_to_zero(speed_df),
                            })

        
        all_tracks.append(description)            
        # Calculate bin properties
        for bin_size in bins:
            bin_end = start_time + bin_size
            s_df = speed_df[speed_df['time'] <= bin_end]
            description = calculate_centroid_properties(s_df)
            description.update({'bid': bid,
                                'eid': eid,
                                'bin-size': bin_size,
                                'id': unique_id,
                                'bl-sprint': fastest_bl_sprint(s_df, typical_bl, max_if_nan=True),
                                'bl-5-sprint': fastest_bl_sprint(s_df, typical_bl*5, max_if_nan=True), 
                                'ac-time-to-0': autocorr_time_to_zero(s_df),
                                })
            
            all_tracks.append(description)            
    df = pd.DataFrame(all_tracks)
    return df.set_index('id')
```
