```python
def show_weirdness(eid, bid):
    print eid, bid
    bid = int(bid)
    sw = SpeedWriter(eid)
    blob = sw._experiment[bid]
    blob_df = sw._pull_blob_positions(blob)
    blob_speed = sw.speed_for_bid(bid)

    print 'raw data'
    plt.plot(blob_df['time'], blob_df['y'], '.')
    plt.plot(blob_df['time'], blob_df['x'], '.')
    plt.xlabel('time')
    
    
    plt.show()
    print 'interpolated/smoothed'
    plt.plot(blob_speed['time'], blob_speed['y'], '.')
    plt.plot(blob_speed['time'], blob_speed['x'], '.')
    plt.xlabel('time')
    plt.show()
    plt.plot(blob_speed['time'], blob_speed['speed'], '.')
    plt.xlabel('time')
    plt.show()
    
    fig, ax = plt.subplots()
    print 'raw x y'
    ax.plot(blob_df['x'], blob_df['y'], '.', label='raw')
    print 'final x y'
    ax.plot(blob_speed['x'], blob_speed['y'], '.', label='final')
    plt.legend(loc='best')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    #return blob_df, blob_speed
```
