```python

def pull_all_xy_for_ex_id(ex_id):
    """
    returns dicts containing times and xy data. keys are blob_ids.
    """
    print(ex_id)
    blob_ids = get_good_blobs(ex_id=ex_id, key='xy_raw')

    print('{N} blob_ids found'.format(N=len(blob_ids)))
    blob_times, blob_xys = {}, {}
    for blob_id in blob_ids:
        times, data = get_timeseries(blob_id, data_type='xy_raw')
        if times != None and len(times):
            blob_times[blob_id] = times
            blob_xys[blob_id] = fill_gaps(data)
    return blob_times, blob_xys

def plot_tracks(blob_xys, max_blobs=50, platesize=[1728, 2352]):
    """ quick dirty plot showing many tracks on plate. 
    
    blob_xys: (dict)
        keys are irrelivant, values are a list of xy tuples.
    max_blobs: (int)
        only plot up to this number of blobs.
    """
    fig, ax = plt.subplots()
    xmax, ymax = 0, 0
    for i, (bid, xy) in enumerate(blob_xys.iteritems()):
        xy = xy[::10]
        #print len(xy)
        x, y = zip(*xy)
        ax.plot(x, y) 
        if max(x) > xmax:
            xmax = max(x)
        if max(y) > ymax:
            ymax = max(y)            
        if i > max_blobs:
            break


    #ax.set_xlim([0, platesize[0]])
    #ax.set_ylim([0, platesize[1]])
    #plt.axis('equal')
    plt.axes().set_aspect('equal', 'datalim')
    try:
        plt.show()
    except OverflowError as e:
        print('Dataset too big to plot')
        print('error:', e)    

    
    
        
ex_id = '20130318_131111'
#ex_id = '20130614_120518'
#ex_id = '20130414_140704'
plate_size = [1728, 2352]

blob_times, blob_xys = pull_all_xy_for_ex_id(ex_id)
plot_tracks(blob_xys, max_blobs=10)
```

```python
def network_report(digraph, nodes=None, verbose=True):
    if nodes is None:
        nodes = digraph

    idig1, idig2, idig3 = itertools.tee(nodes, 3)

    loners = set()
    
    origins = set()
    fission_found = set()
    fusion_found = set()
    
    terminals = set()  
    fission_lost = set()
    fusion_lost = set()
    
    for count, node, (_, in_deg), (_, out_deg) in zip(
            itertools.count(start=1), 
            idig1, 
            dig.in_degree_iter(idig2), 
            dig.out_degree_iter(idig3)):
        
        # finding
        if in_deg == 0:
            origins.add(node)
        elif in_deg == 1:
            fission_found.add(node)
        else: # in_deg > 1
            fusion_found.add(node)

        # losing
        if out_deg == 0:
            terminals.add(node)
        elif out_deg == 1:
            fusion_lost.add(node)
        else: # out_deg > 1
            fission_lost.add(node)
      
    collision_born = fusion_found | fission_found
    collision_lost = fusion_lost | fission_lost    

    if verbose:
        print('Total blobs:       {:6d}'.format(count))
        print('collisison born:   {:6d}'.format(len(collision_born)))
        print('collision lost:    {:6d}'.format(len(collision_lost)))    
        print('0-degree blobs:    {:6d}'.format(len(origins & terminals)))

        print('-'*30)
        print('Found')        
        print('Aether-born blobs: {:6d}'.format(len(origins)))
        print('Fusion-born:       {:6d}'.format(len(fusion_found))) 
        print('Fission-born:      {:6d}'.format(len(fission_found)))
        print('Total:             {:6d}'.format(len(origins | fusion_found | fission_found))) 
        
        print('-'*30)
        print('Lost')        
        print('Abducted blobs:    {:6d}'.format(len(terminals)))
        print('Fusion-lost:       {:6d}'.format(len(fusion_lost))) 
        print('Fission-lost:      {:6d}'.format(len(fission_lost)))
        print('Total:             {:6d}'.format(len(terminals | fusion_lost | fission_lost))) 

    return collision_born, collision_lost

experiment = multiworm.Experiment(where.where(ex_id))
experiment.load_summary(graph=True)
dig = experiment.collision_graph
processed = wio2.Experiment(ex_id)
processed_ids = [worm_id for (worm_id, data) in processed]
print('all\n')
network_report(dig)
print('\n\nprocessed\n')
cborn, clost = network_report(dig, iter(processed_ids))
cborn = ['{eid}_{bid:0=5d}'.format(eid=ex_id, bid=i) for i in cborn]
clost = ['{eid}_{bid:0=5d}'.format(eid=ex_id, bid=i) for i in clost]
```

```python
def plot_exist(fig, ax, blobs, sort_col='t0', color_col='found', first=0, last=3600):
    fill_choice = {'edge':0.5, 'time':1, 'collision':-0.5}
    blobs = blobs.sort(sort_col)
    N_times = last - first
    N_blobs = len(blobs)
    mesh = np.zeros((N_blobs, N_times))
    for i, (bid, row) in enumerate(blobs.iterrows()):
        start, end = int(row['t0']), int(row['tN'])
        fill = fill_choice.get(row[color_col], -1)
        mesh[i][start:end] = fill
        
    mesh[0, 0] = 1.0
    mesh[0, -1] = -1.0
    ppl.pcolormesh(fig, ax, mesh)

print('edge = 0.5')
print('time = 1')
print('collision = -0.5')
print('unknown = -1')
    
fig, ax = plt.subplots()     
plot_exist(fig, ax, blobs, color_col='found')
ax.set_title('found')
plt.show()

fig, ax = plt.subplots()     
plot_exist(fig, ax, blobs, sort_col='tN', color_col='lost')
ax.set_title('lost')
plt.show()
```
