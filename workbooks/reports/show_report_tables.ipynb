# Track Fragmentation

```python
def show_track_termination_report(experiment):
    df = experiment.prepdata.load('end_report')
    df['lifespan'] = ['0-1 min', '1-5 min', '6-10 min', '11-20 min', '21-60 min', 'total']
    df = df.rename(columns={'lifespan': 'track-duration', 
                            'unknown': 'disappear', 
                            'timing':'recording-finishes', 
                            'on_edge':'image-edge'})
    df.set_index('track-duration', inplace=True)
    return df

print 'Reasons for Track Termination'
print
df = show_track_termination_report(experiment)
df
```

```python
def show_track_initiation_report(experiment):
    df = experiment.prepdata.load('start_report')
    df['lifespan'] = ['0-1 min', '1-5 min', '6-10 min', '11-20 min', '21-60 min', 'total']
    df = df.rename(columns={'lifespan': 'track-duration',
                            'unknown': 'appear', 
                            'timing':'recording-begins', 
                            'on_edge':'image-edge'})
    df.set_index('track-duration', inplace=True)
    return df

print 'Reasons for Track Initiation'
print
show_track_initiation_report(experiment)
```
